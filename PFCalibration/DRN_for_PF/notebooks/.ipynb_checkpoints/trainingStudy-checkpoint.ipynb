This is the notebook to open and study the fake and real photon eta and energy pickles in order to derive weights

Also, I'm running this in the Python 3.6.3 Kernel

```python
#import the stuff
import pandas as pd #dataframes etc
import matplotlib.pyplot as plt #plotting
import pickle
import numpy as np
import os, sys
import seaborn as sns
```

```python
#list the pickles
trainingDir = "/home/rusack/evans908/FAIR_GNN/training_folder_maxlr_3e-04_trainbatches_5000_nepoch_50/"
predFile = "pred.pickle"

labelsDir = "/home/rusack/shared/pickles/2018_Gamma_Jet/combPhotons/"
labelsFile = "labels_target.pickle"

ptFile = "pt.pickle"
etaFile = "eta.pickle"

predPickle = open(trainingDir+predFile, "rb")
labelPickle = open(labelsDir+labelsFile, "rb")
ptPickle = open(labelsDir+ptFile, "rb")
etaPickle = open(labelsDir+etaFile, "rb")

preds = pickle.load(predPickle)
labels = pickle.load(labelPickle)
etas = pickle.load(etaPickle)
pts = pickle.load(ptPickle)
```

```python
print(len(preds))
print(len(labels))
print(len(etas))
print(len(pts))
```

```python
res_df = pd.DataFrame(list(zip(preds,labels,etas,pts)), columns=['prediction','target','pt','eta'])
```

```python
res_df.prediction
```

```python
def sigmoid(X):
    if(X > 100):
        return 1.0
    if(X < -100):
        return 0.0
    return 1/(1+np.exp(-X))
```

```python
sigmoid(2.1)
```

```python
res_df['prediction_activated'] = res_df['prediction'].apply(sigmoid)
```

```python
res_grps = res_df.groupby('target')
```

```python
real_preds = res_grps.get_group(1).prediction_activated
```

```python
fake_preds = res_grps.get_group(0).prediction_activated
```

```python
# Density Plot
sns.distplot(real_preds, hist=True, kde=True, 
             bins=100, color = 'darkblue', label = 'Real Photons',
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4}
           )
ax = sns.distplot(fake_preds, hist=True, kde=True, 
             bins=100, color = 'red', label = 'Fake Photons',
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4}
                )

ax.legend(loc='center',facecolor='white',framealpha=1)
ax.set_xlabel('Predicted Probability')
ax.get_yaxis().set_visible(False)
```

```python
n_real = real_preds.size
n_fake = fake_preds.size

real_preds.name = 'Real Photons: '+str(n_real)
fake_preds.name = 'Fake Photons: '+str(n_fake)
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
res_df.shape[0]
```

```python
#super naive guess of the majority class (fake photons)
ns_probs = [0 for _ in range(res_df.shape[0])]
```

```python
drn_fpr, drn_tpr, _ = roc_curve(res_df.target, res_df.prediction_activated, pos_label=1)
ns_fpr, ns_tpr, _ = roc_curve(res_df.target, ns_probs, pos_label=1)

drn_score = roc_auc_score(res_df.target, res_df.prediction_activated)
ns_score = roc_auc_score(res_df.target, ns_probs)
```

```python
# plot the roc curve for the model
fig, ax = plt.subplots(figsize=(15,10))
ax.plot(drn_fpr, drn_tpr, marker='.', label='DRN: {:.3f}'.format(float(drn_score)))
ax.plot(ns_fpr, ns_tpr, linestyle='--', label='Guess Fake: {:.3f}'.format(float(ns_score)))
# axis labels
ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')
# show the legend
fig.legend(loc=(.6,.4))
# show the plot
fig.show()
```
